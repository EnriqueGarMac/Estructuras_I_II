```python

import numpy as np
from anastruct import SystemElements
import collections
collections.Iterable = collections.abc.Iterable

# Crear objeto para la estructura
ss = SystemElements()

# En el módulo de rigidez EA dejo el módulo elástico como E=1. De este modo, al darnos el módulo
# E como incógnita, sé que el resultado tendré sencillamente que dividirlo por el
# módulo E.

# Barras
ss.add_truss_element(location=[[0.,0.], [3,0,0.]],EA=0.1)
ss.add_truss_element(location=[[3.,0.], [6.0,3.]],EA=0.3)
ss.add_truss_element(location=[[0.,3.], [3.,3.]],EA=0.2)
ss.add_truss_element(location=[[3.,3.], [6.,3.]],EA=0.2)
ss.add_truss_element(location=[[0.,0.], [0.,3.]],EA=0.1)
ss.add_truss_element(location=[[0.,0.], [3.,3.]],EA=0.5)
ss.add_truss_element(location=[[3.,0.], [3.0,3.]],EA=0.5)

# Añadimos carrito en 3
ss.add_support_roll(node_id=3, direction=2)
# Añadimos apoyo articulado en 4
ss.add_support_hinged(node_id=1)

# Añadimos cargas puntuales
ss.point_load(4, Fy=-10.0)
ss.point_load(5, Fy=-10.0)
```

```python
# Mostramos estructura generada
ss.show_structure()
```

```python
# Resolvemos la estructura
ss.solve();
```

```python
# Mostramos las reacciones
ss.show_reaction_force()
```

```python
# Mostramos axiles
ss.show_axial_force()

```

```python
# Mostramos deformada
ss.show_displacement()
```
