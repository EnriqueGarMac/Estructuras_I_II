```python

import numpy as np
from anastruct import SystemElements
import collections
collections.Iterable = collections.abc.Iterable

# Crear objeto para la estructura
ss = SystemElements()

EAbars = 63.0E+3  # Area

# Barras
ss.add_truss_element(location=[[0.,0.], [2.3,1.]],EA=EAbars)
ss.add_truss_element(location=[[0.,0.], [2.3,3.]],EA=EAbars)
ss.add_truss_element(location=[[2.3,1.], [2.3,3.]],EA=EAbars)

# A침adimos r칩tula al nodo A
ss.add_support_roll(node_id=1, direction=2)
# A침adimos carrito al nodo B
ss.add_support_hinged(node_id=3)

# A침adimos carga puntual al nodo 2
ss.point_load(2, Fx=22.0)
```

```python
# Mostramos estructura generada
ss.show_structure()
```

```python
# Resolvemos la estructura
ss.solve();
```

```python
# Mostramos las reacciones
ss.show_reaction_force()
```

```python
# Mostramos axiles
ss.show_axial_force()

```

```python
# Mostramos deformada
ss.show_displacement()
```
