```python
# Cargamos el software anastruct
from anastruct import SystemElements

# Cargamos el objeto para las viguetas
ss = SystemElements()
```

```python
# Construcción de la estructura
# Añadimos elemento barra 1
ss.add_element(location=[[0, 0], [0, 4.0]]);
# Añadimos elemento barra 3
ss.add_element(location=[[-1, 4.0], [0, 4.0]]);
# Añadimos elemento barra 4
ss.add_element(location=[[0, 4.0], [2.0, 4.0]], spring={2: 0});
# Añadimos elemento barra 5
ss.add_element(location=[[2.0, 4.0], [3.0, 4.0]]);
# Añadimos elemento barra 6
ss.add_element(location=[[3.0, 4.0], [3.0, 1.0]]);
```

```python
# Añadimos apoyo fijo al nodo 1
ss.add_support_fixed(node_id=1)
# Añadimos carrito al nodo 3
ss.add_support_roll(node_id=6, direction=2)
```

```python
# Añadimos carga triangular
ss.q_load(element_id=1, q=(6,0))
```

```python
# Añadimos carga uniformemente distribuida
ss.q_load(element_id=2, q=(3,3))
ss.q_load(element_id=3, q=(3,3))
ss.q_load(element_id=4, q=(3,3))
ss.q_load(element_id=5, q=(-2,-2))
```

```python
# Mostramos estructura generada
ss.show_structure(title='Ejercicio Pórtico plano')
```

```python
# Resolvemos la estructura
ss.solve();
```

```python
# Mostramos las reacciones
ss.show_reaction_force()
```

```python
# Mostramos axiles
ss.show_axial_force()
```

```python
# Mostramos cortantes
ss.show_shear_force()
```

```python
# Mostramos flectores
ss.show_bending_moment()
```

```python
# Mostramos deformada
ss.show_displacement()
```
