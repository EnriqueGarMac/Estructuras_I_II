```python
from anastruct import SystemElements
import numpy as np

ss = SystemElements()

H = 3.0
H2 = 2.0
L = 3.0
P = 10
ang = 60.0*np.pi/180.0


# Añadimos elemento barra 1
ss.add_element(location=[[0, 0], [0.0, H]], spring={2: 0})
# Añadimos elemento barra 2
ss.add_element(location=[[0.0, H], [L, H]], spring={2: 0})
# Añadimos elemento barra 3
ss.add_element(location=[[L, H], [L, H-H2]])

# Apoyos
ss.add_support_fixed(node_id=4)
ss.add_support_hinged(node_id=1)

# Cargas
ss.q_load(element_id=1, q=(4,2))
ss.q_load(element_id=2, q=(2,2))
ss.point_load(3, Fx=P*np.sin(ang), Fy=-P*np.cos(ang))

# Mostramos estructura generada
ss.show_structure(title='Eliminatorio 22/23')
```

```python
# Resolvemos la estructura
ss.solve()
```

```python
# Mostramos las reacciones
ss.show_reaction_force()
```

```python
# Mostramos cortantes
ss.show_shear_force()
```

```python
# Mostramos flectores
ss.show_bending_moment()
```

```python
# Mostramos flectores
ss.show_axial_force()
```
